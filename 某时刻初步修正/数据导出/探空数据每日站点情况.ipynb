```python
import math
import numpy as np
from numpy import *
import pandas as pd
from pandas import DataFrame
```

```python
data=pd.read_csv("D:/20年海洋/数据/探空数据/12个月全部站点信息/ob_info_new.csv",encoding='gbk')
time=data['时间'].unique()
len(time),time
```

```python
def dayzhandian(s):
    data1=data.loc[data['时间']==s]
    suoyin=np.array(data1['站号'])
    data2=data1.set_index(['站号'])
    a = set([45004, 48407, 48820, 59265, 59431, 59644, 59663, 59758, 59792,59981, 96147, 96315, 96441, 96471, 96481, 96509, 98223, 98328,98646, 48855, 48900, 59280, 59293, 98444, 98618, 56985, 59316,98433, 59948])
    b = set(suoyin)
    c = b-(a & b)
    data3=data2.drop(c,axis=0) 
    return data3
```

```python
for i in range(712):
    print(i)
    d=time[i]
    e=dayzhandian(d)
    f=d.replace(':','')
    g=f.replace('00','')
    e.to_csv("D:/20年海洋/数据/探空数据/分月站点信息/"+str(g)+".csv", encoding = "utf-8", header=True, index=True)
```
