```python
import numpy as np
import pandas as pd
import xgboost
from xgboost import XGBRegressor
```

```python
def shuju(a):
    data=pd.read_csv("D:/20年海洋/3000m以下/数据预处理/神经网络/"+str(a)+".csv")
    data1 = pd.DataFrame(data)
    monie=data1.iloc[:,4]
    train_label_e = data1.iloc[:,5]
    train_feature_e = data1.iloc[:, [1,2,3,4]]
    leng=len(monie)
    moniTem=data1.iloc[:,6]
    train_label_Tem = data1.iloc[:,7]
    train_feature_Tem = data1.iloc[:, [1,2,3,6]]

    monipressure=data1.iloc[:,8]
    train_label_pressure = data1.iloc[:,9]
    train_feature_pressure = data1.iloc[:, [1,2,3,8]]
    return monie,train_label_e,train_feature_e,moniTem,train_label_Tem,train_feature_Tem,monipressure,train_label_pressure,train_feature_pressure
```

```python
def xunlian(moni,train_feature,train_label):
    from sklearn.model_selection import train_test_split
    x_train,x_test,y_train,y_test=train_test_split(train_feature,train_label,test_size=0.2,random_state=6)
    moni_test=[]
    for j in range(0,int(len(train_label)/5)):
        a=x_test.index.values[j]
        moni_test.append(moni[a])
    model = XGBRegressor(n_estimators=697,learning_rate=0.17)
    model.fit(x_train,y_train,eval_set=[(x_train, y_train),(x_test, y_test)],eval_metric='rmse' ,verbose='FALSE')
    pred = model.predict(x_test)
    score1 = model.score(x_train,y_train)
    score = model.score(x_test,y_test)
    print(score1)
    print(score)
    a = 0
    b=0
    print(pred.shape[0])
    for i in range(pred.shape[0]):
        print('pred:', pred[i], 'label:', y_test[x_test.index.values[i]],'moni:',moni_test[i])
        a+=np.abs(pred[i]-y_test[x_test.index.values[i]])
        b+=np.abs(y_test[x_test.index.values[i]]-moni_test[i])
    a=a/(len(train_label)/5)
    b=b/(len(train_label)/5)
    print(a,b)
    return model
```

```python
#时间
a='20170101_00'
monie,train_label_e,train_feature_e,moniTem,train_label_Tem,train_feature_Tem,monipressure,train_label_pressure,train_feature_pressure=shuju(a)
model1=xunlian(monie,train_feature_e,train_label_e)
model2=xunlian(moniTem,train_feature_Tem,train_label_Tem)
model3=xunlian(monipressure,train_feature_pressure,train_label_pressure)
#时间
data1=pd.read_csv("D:/20年海洋/数据/模拟数据/3000m层数层面导出数据/20170101nt1.csv" , encoding = "utf-8")
data2 = pd.DataFrame(data1)

x=data2.iloc[:,[4,5,8,9]]
x=x.set_axis(['lon', 'lat', 'z', 'monie'], axis='columns', inplace=False)
xiuzhenge = model1.predict(x)

x=data2.iloc[:,[4,5,8,10]]
x=x.set_axis(['lon', 'lat', 'z', 'monipressure'], axis='columns', inplace=False)
xiuzhengpressure = model3.predict(x)

x=data2.iloc[:,[4,5,8,11]]
x=x.set_axis(['lon', 'lat', 'z', 'moniTem'], axis='columns', inplace=False)
xiuzhengTem = model2.predict(x)
xiuzhenge,xiuzhengpressure,xiuzhengTem
```
