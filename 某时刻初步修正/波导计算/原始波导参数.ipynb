```python
writer=pd.ExcelWriter("D:/20年海洋/3000m以下/原始波导参数/20170101_00模拟波导数据.xlsx")
for i in range(1,94):
    data_i=data.loc[data['ny']==i]
    a=['波导厚度 m']
    a=pd.DataFrame(a)
    a.to_excel(writer,'ny'+str(i),startrow=1,startcol=0,header=None,index=None)
   
    a=['波导所在底部的海拔 m']
    a=pd.DataFrame(a)
    a.to_excel(writer,'ny'+str(i),startrow=9,startcol=0,header=None,index=None)
    
    a=['波导所在顶部的海拔 m']
    a=pd.DataFrame(a)
    a.to_excel(writer,'ny'+str(i),startrow=17,startcol=0,header=None,index=None)
   
    a=['波导所在顶部的温度 K']
    a=pd.DataFrame(a)
    a.to_excel(writer,'ny'+str(i),startrow=25,startcol=0,header=None,index=None)
    
    a=['波导所在顶部的水汽压e hpa']
    a=pd.DataFrame(a)
    a.to_excel(writer,'ny'+str(i),startrow=33,startcol=0,header=None,index=None)
   
    #a=['波导所在顶部水汽混合比']
    #a=pd.DataFrame(a)
    #a.to_excel(writer,'ny'+str(i),startrow=25,startcol=0,header=None,index=None)
    
    a=['相应饱和水汽压E hpa']
    a=pd.DataFrame(a)
    a.to_excel(writer,'ny'+str(i),startrow=41,startcol=0,header=None,index=None)
   
    a=['相对湿度f']
    a=pd.DataFrame(a)
    a.to_excel(writer,'ny'+str(i),startrow=49,startcol=0,header=None,index=None)
    
    a=['波导所在顶部的气压  hpa（mbar）']
    a=pd.DataFrame(a)
    a.to_excel(writer,'ny'+str(i),startrow=57,startcol=0,header=None,index=None)
    
    for j in range(1,91):
        data_j=data_i.loc[data_i['nx']==j]
        e=array(data_j['e'])
        Tem=array(data_j['Tem'])
        Pressure=array(data_j['pressure'])
        height=array(data_j['z'])
        
        m=77.6*Pressure/Tem+3.73*10**5*e/Tem**2+0.157*height
       
        greater=m[signal.argrelextrema(m, np.greater)]
        print (greater)
        greaterpoint=signal.argrelextrema(m, np.greater)
        print(greaterpoint)
        
        less=m[signal.argrelextrema(m, np.less)]
        print(less)
        lesspoint=signal.argrelextrema(m, np.less)
        print(lesspoint)
        
        if greater.shape==less.shape:
            mdifference=greater-less
        else:
            less=less[1:]
            mdifference=greater-less
            
        Height=height[lesspoint]
        HEIGHT=height[greaterpoint]
        
        if Height.shape==HEIGHT.shape:
            heightdifference=Height-HEIGHT
            toppressure=Pressure[lesspoint]
            toptem=Tem[lesspoint]
            x=53.67957-6743.769/toptem-4.8451*np.log(toptem)
            E=np.exp(x)
            tope=e[lesspoint]
            f=tope/E
        else:
            Height=Height[1:]
            heightdifference=Height-HEIGHT
            toppressure=Pressure[lesspoint]
            toppressure=toppressure[1:]
            toptem=Tem[lesspoint]
            toptem=toptem[1:]
            x=53.67957-6743.769/toptem-4.8451*np.log(toptem)
            E=np.exp(x)
            tope=e[lesspoint]
            tope=tope[1:]
            f=tope/E        
        
        heightdifference=pd.DataFrame(heightdifference)
        heightdifference.to_excel(writer,'ny'+str(i),startrow=1,startcol=j,header=None,index=None)
        HEIGHT=pd.DataFrame(HEIGHT)
        HEIGHT.to_excel(writer,'ny'+str(i),startrow=9,startcol=j,header=None,index=None)
        Height=pd.DataFrame(Height)
        Height.to_excel(writer,'ny'+str(i),startrow=17,startcol=j,header=None,index=None)
        toptem=pd.DataFrame(toptem)
        toptem.to_excel(writer,'ny'+str(i),startrow=25,startcol=j,header=None,index=None)
        tope=pd.DataFrame(tope)
        tope.to_excel(writer,'ny'+str(i),startrow=33,startcol=j,header=None,index=None)
        E=pd.DataFrame(E)
        E.to_excel(writer,'ny'+str(i),startrow=41,startcol=j,header=None,index=None)
        f=pd.DataFrame(f)
        f.to_excel(writer,'ny'+str(i),startrow=49,startcol=j,header=None,index=None)
        toppressure=pd.DataFrame(toppressure)
        toppressure.to_excel(writer,'ny'+str(i),startrow=57,startcol=j,header=None,index=None)
writer.save()
```
